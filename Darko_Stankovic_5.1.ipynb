# Will a Customer Accept the Coupon?

**Context**

Imagine driving through town and a coupon is delivered to your cell phone for a restaraunt near where you are driving. Would you accept that coupon and take a short detour to the restaraunt? Would you accept the coupon but use it on a sunbsequent trip? Would you ignore the coupon entirely? What if the coupon was for a bar instead of a restaraunt? What about a coffee house? Would you accept a bar coupon with a minor passenger in the car? What about if it was just you and your partner in the car? Would weather impact the rate of acceptance? What about the time of day?

Obviously, proximity to the business is a factor on whether the coupon is delivered to the driver or not, but what are the factors that determine whether a driver accepts the coupon once it is delivered to them? How would you determine whether a driver is likely to accept a coupon?

**Overview**

The goal of this project is to use what you know about visualizations and probability distributions to distinguish between customers who accepted a driving coupon versus those that did not.

**Data**

This data comes to us from the UCI Machine Learning repository and was collected via a survey on Amazon Mechanical Turk. The survey describes different driving scenarios including the destination, current time, weather, passenger, etc., and then ask the person whether he will accept the coupon if he is the driver. Answers that the user will drive there ‘right away’ or ‘later before the coupon expires’ are labeled as ‘Y = 1’ and answers ‘no, I do not want the coupon’ are labeled as ‘Y = 0’.  There are five different types of coupons -- less expensive restaurants (under \\$20), coffee houses, carry out & take away, bar, and more expensive restaurants (\\$20 - \\$50). 

**Deliverables**

Your final product should be a brief report that highlights the differences between customers who did and did not accept the coupons.  To explore the data you will utilize your knowledge of plotting, statistical summaries, and visualization using Python. You will publish your findings in a public facing github repository as your first portfolio piece. 





## Data Description

The attributes of this data set include:
1. User attributes
    -  Gender: male, female
    -  Age: below 21, 21 to 25, 26 to 30, etc.
    -  Marital Status: single, married partner, unmarried partner, or widowed
    -  Number of children: 0, 1, or more than 1
    -  Education: high school, bachelors degree, associates degree, or graduate degree
    -  Occupation: architecture & engineering, business & financial, etc.
    -  Annual income: less than \\$12500, \\$12500 - \\$24999, \\$25000 - \\$37499, etc.
    -  Number of times that he/she goes to a bar: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    -  Number of times that he/she buys takeaway food: 0, less than 1, 1 to 3, 4 to 8 or greater
    than 8
    -  Number of times that he/she goes to a coffee house: 0, less than 1, 1 to 3, 4 to 8 or
    greater than 8
    -  Number of times that he/she eats at a restaurant with average expense less than \\$20 per
    person: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    -  Number of times that he/she goes to a bar: 0, less than 1, 1 to 3, 4 to 8 or greater than 8
    

2. Contextual attributes
    - Driving destination: home, work, or no urgent destination
    - Location of user, coupon and destination: we provide a map to show the geographical
    location of the user, destination, and the venue, and we mark the distance between each
    two places with time of driving. The user can see whether the venue is in the same
    direction as the destination.
    - Weather: sunny, rainy, or snowy
    - Temperature: 30F, 55F, or 80F
    - Time: 10AM, 2PM, or 6PM
    - Passenger: alone, partner, kid(s), or friend(s)


3. Coupon attributes
    - time before it expires: 2 hours or one day

# General analysis of the overall dataset




## Read in the `coupons.csv` file.

Using Pandas library I will import the data from the .csv file

```python
# importing data
data = pd.read_csv('data/coupons.csv')
```

```python
# exploring first 10 rows
data.head(10)
```

## Investigate the dataset for missing or problematic data.

I will first explore the data to check the missing values and check the data type for each column. 

```python
#find columns with missing data
print (data.isnull().sum())
```

The columns car, Bar, CoffeeHouse, CarryAway, RestaurantLessThan20, Restaurant20To50 shows a missing values.


```python
 #Explore types of data in the columns
print(data.info())
```

Additionaly, by checking the data types I realized that last column "Y" is an integer insted of boolian (True/False).

In the next steps I will be cleaning the dataset.

## Decide what to do about your missing data -- drop, replace, other...

The column "car" has too many rows of missing data so decide to keep it in the dataset. On the other hand, lower missing row count is in the colimns "Bar" and "RestaurantsLessThan20" so I will be removing na values from the dataset.

```python
#removing the na vales from dataset column "Bar"
data = data[data['Bar'].notna()]

#removing the na vales from dataset column "RestaurantLessThan20"
data = data[data['RestaurantLessThan20'].notna()]

# checking data cleaning
print (data.isnull().sum())
```

Changining column "Y" to Boolean (True/False) to help in the future analysis

```python
#Changeing the "Y" column to Boolean (True/False) 
data['Y'] = data['Y'].astype('bool')

#Check the "Y" column type
print(data.info())
```

## What proportion of the total observations chose to accept the coupon? 

By deviding total count of provided coupons with the number of used coupns I calculated the coupon acceptance  proportion (round to 2 decimal points).

```python
print (round(
    ((data['Y'].values.sum())/(data['Y'].count())*100),2), 
       "% of all coupons are accepted")
```

## Use a bar plot to visualize the `coupon` column.

```python
plt.figure(figsize = (15,8))
fig =sns.countplot(data=data, x='coupon', hue = 'Y',palette=['red', 'lightblue'])
fig.set( title='Coupon acceptance by type')
plt.xlabel("Type of establishment")
plt.ylabel("Count of coupons")
```

The above visualization shows that the most coupons were issued for the Cofee House. However, smaller restaurants and Carry out & Take away users are the ones using the coupons at the higher level compared to Coffe House, Bars and larger restaurants users.

## Use a histogram to visualize the temperature column.

```python
plt.figure(figsize = (15,8))
fig=sns.displot(data=data, x='temperature', hue ="Y",palette=['red', 'lightblue'])
fig.set( title='Coupon acceptance and temperature ')
plt.xlabel("Temperature group")
plt.ylabel("Count of coupons")
```

The above graph shows that there is a correleation between the Temperature and coupon usage. There is a higher coupon usages if the tepmeratures are higher (potentialy during the summer season or places on the south).

# Analysis of the Bar Coupons

Now, I will lead you through an exploration of just the bar related coupons.  


## Create a new `DataFrame` that contains just the bar coupons.

```python
# Filter only Bar coupons and explore the new bar_df dataset
bar_df = data[data['coupon'].str.contains("Bar")]
bar_df.head()
```

## What proportion of bar coupons were accepted?

By deviding total bar coupon count with the number of used bar coupns I calculated the bar coupon acceptance proportion (round to 2 decimal points).

```python
print (round(
    ((bar_df['Y'].values.sum())/(bar_df['Y'].count())*100),2), 
       "% of all bar coupons are accepted")
```

## Compare the acceptance rate between those who went to a bar 3 or fewer times a month to those who went more.

First, I am visualizing data for the bar and acceptance by monthly bar visits. After that I will calculate the acceptance rate for different montly visit counts.

```python
# visualization of the coupons acceptance by different monthly visito counts
plt.figure(figsize = (15,8))
fig=sns.countplot(data=bar_df, x='Bar', hue='Y', order = ['never','less1','1~3', '4~8','gt8'],palette=['red', 'lightblue'])
fig.set( title='Coupon acceptance by monthly bar visits')
plt.xlabel("Monthly visits (count)")
plt.ylabel("Count of coupons")
```

```python
# calculation of the accepatnce between two groups (<=3 and >=4 monthly visits)
less_3 = bar_df.query('Bar == ["Never","less1", "1~3"]')
more_3 = bar_df.query('Bar == ["gt8", "4~8"]')

less_3_ppt = less_3['Y'].value_counts(normalize=True).loc[0]*100
more_3_ppt = more_3['Y'].value_counts(normalize=True).loc[0]*100

print(round(less_3_ppt,1),"% of people that visited bar less than 3 times a month used the coupon while only",
      round(more_3_ppt,1),"% people that vistis the bar more than 3 times a month is using the coupon")
```

## Compare the acceptance rate between drivers who go to a bar more than once a month and are over the age of 25 to the all others.  Is there a difference?


```python
# calculation of the accepatnce between two groups (more than one visit and >25 years old vs everyone else)
more_1_25= bar_df.query('Bar == ["1~3","4~8","gt8"] & age > "25"')
more_other = bar_df.query('Bar != ["1~3","4~8","gt8"] & age < "25"')

more_1_25_ppt = more_1_25['Y'].value_counts(normalize=True).loc[0]*100
more_other_ppt = more_other['Y'].value_counts(normalize=True).loc[0]*100

print(round(more_1_25_ppt,1),"% of people older than 25 visit bar more than once a month and used the coupon while", 
      round(more_other_ppt,1),"% people that used coupon visited bar less then once per month")
```

## Use the same process to compare the acceptance rate between drivers who go to bars more than once a month and had passengers that were not a kid and had occupations other than farming, fishing, or forestry. 


```python
# visulaization of the acceptance rate based on passengers
plt.figure(figsize = (15,8))
fig=sns.countplot(data=data, x='passanger', hue='Y',palette=['red', 'lightblue'])
fig.set( title='Coupon acceptance by passenger types')
plt.xlabel("Users group")
plt.ylabel("Count of coupons")
fig.plot()
```

```python
# visualize the coupon acceptance by visitor ocupation
plt.figure(figsize = (15,8))
ax=sns.countplot(data=data, x='occupation', hue='Y',palette=['red', 'lightblue'])
ax.set( title='Coupon acceptance by visitor occupation')
plt.xlabel("Occupation")
plt.ylabel("Count of coupons")
#Rotate lables
ax.set_xticklabels(ax.get_xticklabels(),rotation = -80)
fig.plot()
```

```python
# calculation of the accepatnce for the group
more_1_noKid_occupation= bar_df.query('Bar == ["1~3","4~8","gt8"] & passanger != "Kid(s)" & occupation != ["farming","fishing","forestry"]')


more_1_noKid_occupation_ppt = more_1_noKid_occupation['Y'].value_counts(normalize=True).loc[0]*100


print(round(more_1_noKid_occupation_ppt,1),
      "% of people with the ocupation other than farming, fishing or forestry visit bar more than once a month without kids and used the coupon")     
```

## Compare the acceptance rates between those drivers who:

- go to bars more than once a month, had passengers that were not a kid, and were not widowed *OR*
- go to bars more than once a month and are under the age of 30 *OR*
- go to cheap restaurants more than 4 times a month and income is less than 50K. 



```python
#create DatFrame of drivers who go to bars more than once a month and had passengers that were not a kid 
# and had occupations other than farming, fishing, or forestry. 
bars_nk_df = data[(data['passanger']!="Kid(s)")& (data['occupation']!="Farming Fishing & Forestry")
                  & ~data['Bar'].str.contains("never") & ~data['Bar'].str.contains("less")]

bars_nk_df_ppt = bars_nk_df['Y'].value_counts(normalize=True).loc[0]*100 

print(round(bars_nk_df_ppt,1),
      "% of people with the ocupation other than farming, fishing or forestry visit bar more than once a month without kids and used the coupon")     
```

```python
#create DatFrame of drivers who go to bars more than once a month and older than 25
more_1_under25= bar_df.query('Bar == ["1~3","4~8","gt8"] & age > "25"')

more_1_under25_ppt = more_1_under25['Y'].value_counts(normalize=True).loc[0]*100 

print(round(more_1_under25_ppt,1),
      "% of people older than 25 and visit bar more than once a month used the coupon")     
```

```python
#Calculate acceptance rates
bar_4_ch = data[((data['coupon'].str.contains("Bar")) & (data['income'] <'50000')) & ((data['RestaurantLessThan20'] == '4~8') | (data['RestaurantLessThan20'] == 'gt8'))   ]
Accept = (bar_4_ch['Y'].value_counts())
Total = (bar_4_ch['Y'].count())
print('Given group had,', Accept[0], 'rejected the coupon and', Accept[1], 'accepted','for', round((Accept[1]/Total),2),'% acceptance rate' )
```

```python
plt.figure(figsize = (15,8))
ax=sns.countplot(data=bar_4_ch, x='occupation', hue='Y', palette=['red', 'lightblue'])
ax.set_xticklabels(ax.get_xticklabels(),rotation = -80)
ax.set( title='Coupon acceptance by visitor income')
plt.xlabel("Occupation")
plt.ylabel("Count of coupons")
ax.plot()
```
